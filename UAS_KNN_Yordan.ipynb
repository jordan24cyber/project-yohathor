# **Soal 1 – Load & Eksplorasi Data**

**a. Load Dataset**

```python
import numpy as np
from sklearn.datasets import load_iris

# Load dataset
data = load_iris()
X = data.data
y = data.target
```

**b. Eksplorasi Data**

```python
# Jumlah data dan fitur
print("Jumlah data:", X.shape[0])
print("Jumlah fitur:", X.shape[1])

# Nama fitur
print("Nama fitur:", data.feature_names)

# Distribusi kelas target
unique, counts = np.unique(y, return_counts=True)
print("Distribusi kelas target:")
for u, c in zip(unique, counts):
    print(f"Kelas {u}: {c} data")
```

# **Soal 2 – Preprocessing**

**a. Split Data (70% Train, 30% Test)**

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)
```

**b. Feature Scaling (StandardScaler)**

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

# **Soal 3 – KNN**

**a. Implementasi Model KNN (k = 5)**

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=5)
```

**b. Training & Prediksi**

```python
knn.fit(X_train_scaled, y_train)
y_pred = knn.predict(X_test_scaled)
```

**c. Evaluasi Model**

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='macro')
recall = recall_score(y_test, y_pred, average='macro')
f1 = f1_score(y_test, y_pred, average='macro')

print("Accuracy :", accuracy)
print("Precision:", precision)
print("Recall   :", recall)
print("F1-score :", f1)
```
