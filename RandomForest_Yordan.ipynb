# **PERSIAPAN DATASET**

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

# **MEMBACA DATASET**

```python
df = pd.read_csv('earthquake_alert_balanced_dataset.csv')
```

```python
X = df.drop('alert', axis=1)
y = df['alert']

print(f'Dimensi Feature: {X.shape}')
print(f'Class: {set(y)}')
```

```python
X = df.drop('alert', axis=1)
y = df['alert']

X_train, X_test, y_train, y_test = train_test_split(X, y,
test_size=0.3, random_state=42)
```

# **Classification RandomForest Classifier**

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=100, random_state=0)

model.fit(X_train, y_train)
```

# **Evaluasi Model**

```python
from sklearn.metrics import classification_report

y_pred = model.predict(X_test)

print(classification_report(y_test, y_pred))
```

# **INPUT MANUAL RANDOM FOREST**

```python
import time
import pandas as pd

# --- Input manual ---
print("Masukkan parameter gempa untuk prediksi:")
magnitude = float(input("Magnitude : "))
depth = float(input("Depth : "))
cdi = float(input("CDI : "))
mmi = float(input("MMI : "))
sig = float(input("SIG : "))

# --- Data input diubah ke bentuk DataFrame sesuai urutan kolom dataset ---
input_data = pd.DataFrame([[magnitude, depth, cdi, mmi, sig]],
                          columns=['magnitude', 'depth', 'cdi', 'mmi', 'sig'])

# --- Mulai timing training model (contoh: Naive Bayes) ---
start_time = time.time()
model.fit(X_train, y_train)
training_time = time.time() - start_time

# --- Prediksi data input ---
prediksi = model.predict(input_data)[0]

# --- Evaluasi model pada data test ---
from sklearn.metrics import accuracy_score, precision_score, recall_score

y_pred_test = model.predict(X_test)
akurasi = accuracy_score(y_test, y_pred_test) * 100
precision = precision_score(y_test, y_pred_test, average='weighted')
recall = recall_score(y_test, y_pred_test, average='weighted')

# --- Tampilkan hasil ---
print("\n--- 🔍 Hasil Prediksi ---")
print(f"Hasil Prediksi Alert : {prediksi}")

# --- Tabel ringkasan metrik seperti contoh ---
data_tabel = {
    'Algoritma': ['Random Forest'],
    'Akurasi (%)': [round(akurasi, 2)],
    'Precision': [round(precision, 4)],
    'Recall': [round(recall, 4)],
    'Waktu Training (S)': [round(training_time, 6)]
}

df_tabel = pd.DataFrame(data_tabel)
# --- Cetak tabel metrik ---
print("===========================================================")
print("--- RINGKASAN METRIK MODEL (Data Test) ---")
print("| Algoritma        | Akurasi (%) | Precision | Recall  | Waktu Training (S) |")
print("|------------------|-------------|-----------|---------|--------------------|")
print(f"| Random Forest    | {akurasi:9.2f}   | {precision:8.4f} | {recall:7.4f} | {training_time:18.6f} |")
print("===========================================================\n")
```
